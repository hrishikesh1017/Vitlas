```python
import pandas as pd

# Define the MEWS scoring functions
def assign_mews_score_to_pulse_rate(pulse_rate):
    if pulse_rate == 'N':
        return 'N'
    pulse_rate = float(pulse_rate)
    if 51 <= pulse_rate <= 100:
        return 0
    elif 101 < pulse_rate <= 110:
        return 1
    elif 41 < pulse_rate <= 50:
        return 1
    elif 111 < pulse_rate <= 129:
        return 2
    elif pulse_rate <= 40:
        return 2
    elif pulse_rate >= 130:
        return 3
    else:
        return 3

def assign_mews_score_to_spo2(spo2):
    if spo2 == 'N':
        return 'N'
    spo2 = float(spo2)
    if 95 <= spo2:
        return 0
    elif 90 <= spo2 <= 94:
        return 1
    elif 86 <= spo2 <= 89:
        return 2
    elif spo2 <= 85:
        return 3
    else:
        return 3

def assign_mews_score_to_respiratory_rate(respiratory_rate):
    if respiratory_rate == 'N':
        return 'N'
    respiratory_rate = float(respiratory_rate)
    if 9 <= respiratory_rate <= 14:
        return 0
    elif 15 <= respiratory_rate <= 20:
        return 1
    elif 21 <= respiratory_rate <= 29:
        return 2
    elif 30 <= respiratory_rate:
        return 3
    else:
        return 3

def assign_mews_score_to_temperature(temperature):
    if temperature == 'N':
        return 'N'
    temperature = float(temperature)
    if 35 <= temperature <= 36:
        return 1
    elif 36.1 <= temperature <= 38:
        return 0
    elif 38.1 <= temperature <= 39:
        return 1
    elif temperature > 39:
        return 3
    else:
        return 2

def assign_mews_score_to_bp(bp):
    if bp == 'N':
        return 'N'
    bp = float(bp)
    if bp < 90:
        return 2
    elif 90 <= bp <= 100:
        return 0
    elif 101 <= bp <= 110:
        return 1
    elif bp > 110:
        return 3
    else:
        return 2


df = pd.read_csv("DataSets.csv")

# Replace empty values with 'N'
df = df.fillna('N')

df_mews = pd.DataFrame()

# Apply the MEWS scoring functions
df_mews['Time'] = df["Time"]
df_mews['PR_Val'] = df['PR_Val'].apply(assign_mews_score_to_pulse_rate)
df_mews['Temp_Val'] = df['Temp_Val'].apply(assign_mews_score_to_temperature)
df_mews['Sys_Val'] = df['Sys_Val'].apply(assign_mews_score_to_bp)
df_mews['Resp_Val'] = df['Resp_Val'].apply(assign_mews_score_to_respiratory_rate)
df_mews['SPO2'] = df['SPO2'].apply(assign_mews_score_to_spo2)

print(df_mews)

df_mews.to_csv("test_datasheet1.csv", index=False)
```
